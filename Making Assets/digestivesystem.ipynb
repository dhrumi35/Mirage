```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the digestive system organs
organs = [
    {"name": "Mouth", "location": (0, 0, 0), "scale": (1, 1, 1)},
    {"name": "Esophagus", "location": (0, 2, 0), "scale": (0.5, 1, 0.5)},
    {"name": "Stomach", "location": (0, 4, 0), "scale": (1, 1, 1)},
    {"name": "Small Intestine", "location": (0, 6, 0), "scale": (0.5, 1, 0.5)},
    {"name": "Large Intestine", "location": (0, 8, 0), "scale": (1, 1, 1)},
    {"name": "Rectum", "location": (0, 10, 0), "scale": (0.5, 1, 0.5)},
]

# Create a figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a sphere for each organ
for organ in organs:
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x = organ["scale"][0] * np.cos(u) * np.sin(v) + organ["location"][0]
    y = organ["scale"][1] * np.sin(u) * np.sin(v) + organ["location"][1]
    z = organ["scale"][2] * np.cos(v) + organ["location"][2]
    ax.plot_wireframe(x, y, z, color="b")

# Set axis labels and title
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("3D Digestive System Model")

# Show the plot
plt.show()
```
